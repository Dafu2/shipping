```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv
from datetime import datetime
import time
import matplotlib.dates as mdates
from matplotlib.pyplot import figure


%matplotlib inline
```

```python
print("Starting program.")
```

```python
#DATA - API CALLs
DATA_PATH = '/Users/Xiao/Desktop/warp_gate/shipping/data/FINANCIAL'

def load_financial_data(data_path=DATA_PATH):
    csv_path = os.path.join(data_path, "LME-PR_CO_5Y.csv")
    return pd.read_csv(csv_path)

finance = load_financial_data(DATA_PATH)
# print(finance)
```

```python
# print(finance['Cash Buyer'])
#Plot price vs date
dates = finance['Date']
value_list = finance['Cash Buyer']
# date = []
# for time_val in date_value:
#     date.append(datetime(*time.strptime(time_val, "%Y-%m-%d")[:5]))

date_objects = [datetime.strptime(date, '%Y-%m-%d').date() for date in dates]
    
# print (date_objects)
fig, ax = plt.subplots()

ax.plot_date(date_objects, value_list)
date_format = mdates.DateFormatter("%Y-%m-%d")
ax.xaxis.set_major_formatter(date_format)
ax.autoscale_view()
title_name = 'Cash Buyer vs. Date'
ax.set_title(title_name)
y_label = 'Cash Value'
ax.set_ylabel(y_label)
ax.grid(True)

fig.autofmt_xdate()
plt.show()

```
